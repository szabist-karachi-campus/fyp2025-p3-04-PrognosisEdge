```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the unified dataset
file_path = "C:/Users/ghaza/Desktop/FYP/Application/AI/data/processed/preprocessed_data.csv"
data = pd.read_csv(file_path)

data.head()
```

```python
# Separate features and targets
features = data.drop(columns=["machine_failure", "failure_type_encoded"])
binary_target = data["machine_failure"]
multi_target = data["failure_type_encoded"]
```

```python
# Split the dataset into training and testing sets (70-30 ratio)
X_train, X_test, y_train_binary, y_test_binary = train_test_split(
    features, binary_target, test_size=0.3, random_state=42, stratify=binary_target
)
```

```python
# Split the multi-class target using the same indices as the binary split
_, _, y_train_multi, y_test_multi = train_test_split(
    features, multi_target, test_size=0.3, random_state=42, stratify=binary_target
)
```

```python
# Perform SMOTEENN sampling for binary target
smoteenn_binary = SMOTEENN(random_state=42)
X_train_balanced_binary, y_train_balanced_binary = smoteenn_binary.fit_resample(
    X_train, y_train_binary
)

# Verify the balanced binary dataset
print("\nBinary Classification - Balanced Dataset:")
print("X_train_balanced_binary:", X_train_balanced_binary.shape)
print("y_train_balanced_binary:", y_train_balanced_binary.shape)
print("Class distribution:", pd.Series(y_train_balanced_binary).value_counts())
```

```python
# Combine X_train_balanced_binary and y_train_balanced_binary into a DataFrame
balanced_binary_data = pd.DataFrame(X_train_balanced_binary, columns=features.columns)
balanced_binary_data["machine_failure"] = y_train_balanced_binary

# Merge with the multi-class target
aligned_multi_data = balanced_binary_data.merge(
    pd.concat([X_train, y_train_multi], axis=1),
    how="inner",
    on=X_train.columns.tolist(),
)

# Separate features and multi-class target after alignment
X_train_aligned_multi = aligned_multi_data.drop(columns=["machine_failure", "failure_type_encoded"])
y_train_aligned_multi = aligned_multi_data["failure_type_encoded"]

# Verify alignment
print("\nAligned Multi-Class Training Data:")
print("X_train_aligned_multi:", X_train_aligned_multi.shape)
print("y_train_aligned_multi:", y_train_aligned_multi.shape)
```

```python
# Verify the split
print("Shapes after splitting:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train_binary:", y_train_binary.shape)
print("y_test_binary:", y_test_binary.shape)
print("y_train_multi:", y_train_multi.shape)
print("y_test_multi:", y_test_multi.shape)
```

```python
# Perform SMOTEENN sampling for multi-class target
smoteenn_multi = SMOTEENN(random_state=42)
X_train_balanced_multi, y_train_balanced_multi = smoteenn_multi.fit_resample(
    X_train_aligned_multi, y_train_aligned_multi
)

# Verify the balanced multi-class dataset
print("\nMulti-Class Classification - Balanced Dataset:")
print("X_train_balanced_multi:", X_train_balanced_multi.shape)
print("y_train_balanced_multi:", y_train_balanced_multi.shape)
print("Class distribution:", pd.Series(y_train_balanced_multi).value_counts())
```

```python
def plot_target_distribution(target, title, xlabel, ylabel):
    
    # Calculate class counts
    class_counts = pd.Series(target).value_counts().sort_index()

    # Plot the distribution
    plt.figure(figsize=(10, 6))
    bars = plt.bar(class_counts.index, class_counts.values, color="skyblue", alpha=0.8)

    # Add data labels dynamically above the bars
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2, 
            height + height * 0.02,  # Offset based on bar height (2% higher)
            f"{int(height)}", 
            ha="center", 
            va="bottom", 
            fontsize=10
        )

    # Set title and labels
    plt.title(title, fontsize=14, fontweight="bold")
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
# Plot distribution for binary target after sampling
plot_target_distribution(
    y_train_balanced_binary,
    title="Distribution of Binary Target (Machine Failure) After Sampling",
    xlabel="Machine Failure (0: No Failure, 1: Failure)",
    ylabel="Count"
)
```

```python
# Plot distribution for multi-class target after sampling
plot_target_distribution(
    y_train_balanced_multi,
    title="Distribution of Multi-Class Target (Failure Type) After Sampling",
    xlabel="Failure Type Encoded",
    ylabel="Count"
)
```

```python
# Specify the file path
file_path = "C:/Users/ghaza/Desktop/FYP/Application/AI/data/processed/"

# Save binary datasets
X_train_balanced_binary.to_csv(f"{file_path}balanced_X_train_binary.csv", index=False)
y_train_balanced_binary.to_csv(f"{file_path}balanced_y_train_binary.csv", index=False)

# Save multi-class datasets
X_train_balanced_multi.to_csv(f"{file_path}balanced_X_train_multi.csv", index=False)
y_train_balanced_multi.to_csv(f"{file_path}balanced_y_train_multi.csv", index=False)

# Save test datasets (unchanged)
X_test.to_csv(f"{file_path}X_test.csv", index=False)
y_test_binary.to_csv(f"{file_path}y_test_binary.csv", index=False)
y_test_multi.to_csv(f"{file_path}y_test_multi.csv", index=False)

print("Datasets have been successfully saved!")
```
